```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
def eulerx(v0x=(.02231875)*365,t0=0,x0=.0508011,dt=0.0001,nsteps=100):
    rp=(((0.0508011)**2)+((0.4011642)**2)+((-0.2195576)**2))**(1/2)
    x=[x0]
    t=[t0]
    vx=[v0x]
    for i in range(1,nsteps):
        xi=x0+v0x*dt
        vxi=v0x-((4*x0*dt*(math.pi)**2)/rp**3)
        x.append(xi)
        t.append(t0+dt*i)
        vx.append(vxi)
        x0=xi
        v0x=vxi
    return x,vx,
```

```python
def eulery(vy0=(.00485553)*365,t0=0,y0=-0.4011642,dt=0.0001,nsteps=100):
    rp=(((0.0508011)**2)+((0.4011642)**2)+((-0.2195576)**2))**(1/2)
    y=[y0]
    t=[t0]
    vy=[vy0]
    for i in range(1,nsteps):
        yi=y0+vy0*dt
        vyi=vy0-((4*y0*dt*(math.pi)**2)/rp**3)
        y.append(yi)
        t.append(t0+dt*i)
        vy.append(vyi)
        y0=yi
        vy0=vyi
    return y,vy,
```

```python
x,vx=eulerx(0,1,2,nsteps=1000)
vy,y=eulery(0,1,2,nsteps=100)
plt.plot(x,y)

```
